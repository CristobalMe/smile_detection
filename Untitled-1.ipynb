# Create Target Directories
Create the target directories if they do not exist.

```python
# Create Target Directories

import os

# Define the paths to the source folder and the target folders
source_folder = "smiles/faces/"
non_smile_folder = "faces/non_smile/"
smile_folder = "faces/smile/"

# Create target folders if they don't exist
os.makedirs(non_smile_folder, exist_ok=True)
os.makedirs(smile_folder, exist_ok=True)
```

# Move Images to Target Directories
Iterate through the images in the source folder and move them to the appropriate target directory based on their names.

```python
import shutil

# Move images to the appropriate target directories based on their names
for filename in os.listdir(source_folder):
    print(filename)
    if filename in NON_SMILE_list:
        shutil.move(os.path.join(source_folder, filename), os.path.join(non_smile_folder, filename))
        old_file = os.path.join(non_smile_folder, filename)
        new_filename = filename.replace('.pmm', '.jpg')
        new_file = os.path.join(non_smile_folder, new_filename)
        
        os.rename(old_file, new_file)
    elif filename in SMILE_list:
        shutil.move(os.path.join(source_folder, filename), os.path.join(smile_folder, filename))
        old_file = os.path.join(smile_folder, filename)
        new_filename = filename.replace('.pmm', '.jpg')
        new_file = os.path.join(smile_folder, new_filename)
        
        os.rename(old_file, new_file)
```

```python
import os

# Define the directory containing the .pmm files
directory = 'faces/non_smile/'

# Loop through all files in the directory
for filename in os.listdir(directory):
    # Check if the file has a .pmm extension
    print(filename)
    if filename.endswith('.jpg'):
        # Define the new filename with a .jpg extension
        new_filename = filename.replace('.jpg', '.ppm')
        # Get the full path to the old and new files
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_filename)
        # Rename the file
        os.rename(old_file, new_file)

print("Renaming complete.")
```
